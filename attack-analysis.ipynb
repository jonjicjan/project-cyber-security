```python
import matplotlib.pyplot as plt
import numpy as np

# Data
years = ['2022', '2023', '2024']
dos = [88, 89, 92]
ddos = [72, 99, 55]
injection = [55, 65, 65]

bar_width = 0.25
x = np.arange(len(years))

# Plotting
plt.figure(figsize=(8, 6))
plt.bar(x - bar_width, dos, width=bar_width, label='DoS', color='white', edgecolor='black')
plt.bar(x, ddos, width=bar_width, label='DDoS', color='limegreen')
plt.bar(x + bar_width, injection, width=bar_width, label='Injection', color='blue')

# Styling
plt.xticks(x, years)
plt.yticks(np.arange(0, 110, 25))
plt.ylim(0, 110)
plt.title('Types of Attacks Over Years')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.gca().set_facecolor('black')       # Set background to black
plt.gcf().patch.set_facecolor('black') # Set figure background to black
plt.tick_params(colors='white')        # Set tick colors to white
plt.xlabel('Year', color='white')
plt.ylabel('Number of Attacks', color='white')
plt.title('Attack Types Comparison (2022-2024)', color='white')

plt.show()
```
